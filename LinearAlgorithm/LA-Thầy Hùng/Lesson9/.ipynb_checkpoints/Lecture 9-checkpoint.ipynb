### Khai triá»ƒn SVD

```python
A=np.array([[4,11,14],[8,7,-2]])
print(A)
```

```python
C=A.T@A
print(C)
```

```python
lam,v=np.linalg.eigh(C)
print(lam.round(3))
print(v.round(3))
```

```python
u1=A.dot(v[:,2])/np.sqrt(lam[2])
print(u1)
```

```python
u2=A.dot(v[:,1])/np.sqrt(lam[1])
print(u2)
```

```python
U=np.array([u1,u2])
print(U)
```

```python
S=np.sqrt(lam.round(3))
print(S)
```

```python
lam1=np.sqrt(lam)
```

```python
U
```

```python
V=v.T
V
```

```python
D=np.zeros(A.shape)
```

```python
D[1,1]=np.sqrt(lam1[1])
D[0,0]=np.sqrt(lam1[2])
print(D)
```

```python
U@D@V
```

```python
C=A.dot(A.T) #AA^T
print("AA^T=",C)
```

```python
lam1,u=np.linalg.eigh(C)
print(lam1)
print(u.round(3))
```

```python
v0=A.T.dot(u[:,0])/np.sqrt(lam1[0])
v1=A.T.dot(u[:,1])/np.sqrt(lam1[1])
```

```python
V=np.array([v0,v1])
print(V)
```

```python
S=np.diag(np.sqrt(lam1))
print(S)
```

```python
u.dot(S.dot(V))
```

```python
D=A.T.dot(A)
print(D)
```

```python
lam2,v=np.linalg.eigh(D)
print(lam2.round(3))
print(v.round(3))
```

```python
s1=np.sqrt(lam1[1])
s2=np.sqrt(lam1[0])
Sr=np.diag([s1,s2])
print(Sr)
```

```python
r=np.linalg.matrix_rank(A)
print(r)
```

```python
Ur=np.array([u[:,1],u[:,0]])
print(Ur)
```

```python
Vr=np.array([v[:,2],v[:,1]])
print(Vr)
```

```python
#Ur*Sr*Vr^T
Ur.dot(Sr.dot(Vr))
```

```python
# SVD
u,s,vt=np.linalg.svd(A)
print("u=",u.round(3))
print("s=",s.round(3))
print("vt=",v.round(3))
```

```python
u.dot(np.diag(s).dot(vt[:2]))
```

```python
u.round(3)
```

```python
D=np.zeros(A.shape)
D[0,0]=s[0]
D[1,1]=s[1]
print(D)
```

```python
np.linalg.pinv(A)
```

```python
Ad=v.T[:,:2].dot(np.linalg.inv(D).dot(u.T))
print(Ad)
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#img = Image.open('test.jpg')
img = Image.open('Lenna.png')
```

```python
img.size
```

```python
plt.imshow(img)
```

```python
imggray = img.convert('LA')
```

```python
plt.imshow(imggray)
plt.show()
```

```python
imgmat = np.array( list(imggray.getdata(band = 0)), float)
imgmat.shape = (imggray.size[1], imggray.size[0])
imgmat = np.matrix(imgmat)
```

```python
plt.imshow(imgmat, cmap = 'gray')
plt.savefig('Lenna_gray.png')
plt.show()
```

```python
print("After compression: ")
U, S, Vt = np.linalg.svd(imgmat) #single value decomposition
```

```python
k=5
cmpimg = np.matrix(U[:, :k]) * np.diag(S[:k]) * np.matrix(Vt[:k,:])
plt.imshow(cmpimg, cmap = 'gray')
plt.savefig('test_'+str(k)+'.png')
plt.show()
```

```python
k=10
cmpimg = np.matrix(U[:, :k]) * np.diag(S[:k]) * np.matrix(Vt[:k,:])
plt.imshow(cmpimg, cmap = 'gray')
plt.savefig('test_'+str(k)+'.png')
plt.show()
```

```python
k=25
cmpimg = np.matrix(U[:, :k]) * np.diag(S[:k]) * np.matrix(Vt[:k,:])
plt.imshow(cmpimg, cmap = 'gray')
plt.savefig('test_'+str(k)+'.png')
plt.show()
```

```python
k=50
cmpimg = np.matrix(U[:, :k]) * np.diag(S[:k]) * np.matrix(Vt[:k,:])
plt.imshow(cmpimg, cmap = 'gray')
plt.savefig('test_'+str(k)+'.png')
plt.show()
```
