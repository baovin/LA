## Bài 1

* Giả sử ma trận $A$ có khái triển SVD $A=\sum\limits_{i=1}^r \sigma_iu_iv_i^T$.
* Nghiệm của bào toán hiệu chỉnh thỏa mãn phương trình $$(A^TA+\lambda I)x=A^Tb. $$
* Ta có $A^TA+\lambda I=\sum\limits_{i=1}^r \sigma_i^2v_iv_i^T +\lambda \sum\limits_{i=1}^n v_iv_i^T=\sum\limits_{i=1}^r (\sigma_i^2+\lambda)v_iv_i^T +\lambda \sum\limits_{i=r+1}^n v_iv_i^T$
* Vậy $$(A^TA+\lambda I)^{-1}=\sum\limits_{i=1}^r \dfrac{1}{\sigma_i^2+\lambda}v_iv_i^T +\dfrac{1}{\lambda} \sum\limits_{i=r+1}^n v_iv_i^T $$
* Nghiệm được tính theo công thức
\begin{align}
x&=(A^TA+\lambda I)^{-1}A^Tb\\
&=\sum\limits_{i=1}^r \dfrac{\sigma_i(u_i^Tb)}{\sigma_i^2+\lambda}v_i 
\end{align}

## Bài 3 Phân tích ảnh

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import data
from skimage.color import rgb2gray
```

```python
ori_image=rgb2gray(data.astronaut())
plt.figure(figsize=(8,8))
plt.imshow(ori_image,cmap='gray')
plt.show()
```

```python
def SVD_truncated(image,k=10):
    s,lam,v=np.linalg.svd(image)
    S=s[:,:k]
    D=np.diag(lam[:k])
    V=v[:k,:]
    Ak=np.dot(S,np.dot(D,V))
    plt.figure(figsize=(8,8))
    plt.imshow(Ak,cmap='gray')
    plt.show()
```

```python
k=10
SVD_truncated(ori_image,k)
```

```python
k=50
SVD_truncated(ori_image,k)
```

```python
k=100
SVD_truncated(ori_image,k)
```
